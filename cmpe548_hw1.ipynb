### Question 1


```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline  

N = 2000
x_arr = []
y_arr = []

for i in range(N):
    theta = 2.0*np.pi*np.random.rand(1)
    r = np.sqrt(np.random.rand(1)) #inversion
    x_arr.append(r*np.cos(theta))
    y_arr.append(r*np.sin(theta))

plt.plot(x_arr, y_arr, 'r.')
plt.axis([-2, 2, -1.5, 1.5])
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

### Question 2

```python
def gatekeeper(x, y, p):
    xx = np.absolute(x)
    yy = np.absolute(y)
    dist = np.power(np.power(xx,p) + np.power(yy,p), np.power(p,-1))

    if dist <= 1:
        return 1 # inside
    else:
        return 0 # outside

p = 1.5
x_arr = []
y_arr = []
acc = 0 # number of accepted points

for i in range(N):
    theta = 2.0*np.pi*np.random.rand(1)
    r = np.sqrt(np.random.rand(1))
    x = r*np.cos(theta)
    y = r*np.sin(theta)
    
    if gatekeeper(x, y, p):
        acc += 1
        x_arr.append(x)
        y_arr.append(y)

plt.plot(x_arr, y_arr, 'r.')
plt.axis([-2, 2, -1.5, 1.5])
plt.xlabel('x')
plt.ylabel('y')
plt.title('p=1.5')
plt.show()
print('Acceptance rate :', acc/N)
```

```python
p = 0.7
x_arr = []
y_arr = []
acc = 0 # number of accepted points

for i in range(N):
    theta = 2.0*np.pi*np.random.rand(1)
    r = np.sqrt(np.random.rand(1))
    x = r*np.cos(theta)
    y = r*np.sin(theta)
    
    if gatekeeper(x, y, p):
        acc += 1
        x_arr.append(x)
        y_arr.append(y)

plt.plot(x_arr, y_arr, 'r.')
plt.axis([-2, 2, -1.5, 1.5])
plt.xlabel('x')
plt.ylabel('y')
plt.title('p=0.7')
plt.show()
print('Acceptance rate :', acc/N)
```

```python
# More efficient sampler

def gatekeeper_v2(x, y):
    xx = np.absolute(x)
    yy = np.absolute(y)
    dist = xx + yy

    if dist <= 1:
        return 1 # inside
    else:
        return 0 # outside

acc = 0 # number of accepted points

for i in range(N):
    theta = 2.0*np.pi*np.random.rand(1)
    r = np.sqrt(np.random.rand(1))
    x = r*np.cos(theta)
    y = r*np.sin(theta)

    if gatekeeper_v2(x, y):
        acc += 1
        
print('Acceptance rate :', acc/N)
```
